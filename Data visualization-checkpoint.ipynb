```python
import matplotlib.pyplot as plt
```

```python
age_x=[25,26,27,28,29,30,31,32]
```

```python
salary_y=[38000,40000,45000,50000,48000,46000,67000,92000]
```

```python
sal_z=[41000,52000,63000,54000,65000,76000,87000,88000]
plt.plot(age_x,salary_y,'r--')
plt.plot(age_x,sal_z,'g--')
plt.title("Salary (USD) of web developers")
plt.xlabel("AGE")
plt.ylabel("Salary")
plt.legend(["All Developers","Python Developers"])
plt.grid()
```

```python
import matplotlib.pyplot as plt
marks=[90,80,70,96,95,86,84,56,35,44,67]
plt.xlabel("Marks of students")
plt.ylabel("Count")
#bin=[0,35,80,95]
plt.hist(marks,bins=5,edgecolor="red",color="yellow")
```

```python
x=[11,22,33,44,55]
y=[35,47,57,68,89]
plt.scatter(x,y,s=100,edgecolor="black",color="red",linewidth=2,marker='*',alpha=0.3)
plt.title("simple scatter plot")
```

```python
students=[23,17,35,29]
courses=["PHP","JAVA","C++","REACTJS"]
plt.title("Number of students enrolled in diff courses")
plt.barh(courses,students,color="orange",height=0.4)
```

```python
students=[23,17,35,29]
courses=["PHP","JAVA","C++","REACTJS"]
plt.title("Number of students enrolled in diff courses")
plt.bar(courses,students,width=0.5,color="orange")
```

```python
stu_per=["EXCELLENT","VERY GOOD","GOOD","AVERAGE","POOR"]
stu_val=[15,25,20,25,15]
plt.figure(figsize=(10,12))
plt.pie(stu_val,labels=stu_per,explode=[0.2,0,0,0,0],shadow=True,colors=["blue","grey","brown","yellow","red"],autopct="%0.2f%%")
plt.legend(title="Performances")
```
